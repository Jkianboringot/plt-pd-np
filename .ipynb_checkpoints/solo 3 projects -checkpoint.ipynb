```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
sale=pd.read_csv("sales_data.csv")
sale.head()
```

## TASK 2 沽

```python
Headphones=sale[sale["product"] == "Headphones"][["total_sales"]]
Monitor=sale[sale["product"] == "Monitor"][["total_sales"]]
Laptop=sale[sale["product"] == "Laptop"][["total_sales"]]
Smartwatch =sale[sale["product"] == "Smartwatch"][["total_sales"]]
tablet=sale[sale["product"] == "Tablet"][["total_sales"]]
```

```python
products=[Headphones.sum().iloc[0],Monitor.sum().iloc[0],Laptop.sum().iloc[0],Smartwatch.sum().iloc[0],tablet.sum().iloc[0]]
label=["Headphones","Monitor","Laptop","Smartwatch","tablet"]
plt.figure(figsize=[8,3])
plt.title("Total sales per product")
plt.bar(label,products)
plt.xlabel("Product")
plt.ylabel("product total_sales")
plt.show()
```

```python
sale['order_date']=pd.to_datetime(sale['order_date'])
sale["monthly_sales"]=sale["order_date"].dt.to_period('M')
k=sale.groupby("monthly_sales")["total_sales"].sum().reset_index()
k["monthly_sales"] = k["monthly_sales"].astype(str) 
```

```python
plt.figure(figsize=[11,10])

plt.title('Monthly sales trends')

plt.plot(k["monthly_sales"],k["total_sales"],"b.-")
plt.xlabel("date")
plt.ylabel("monthly total_sales")
plt.show()
```

```python
region_sales=sale.groupby("region")["total_sales"].sum().reset_index()

plt.title(" Sales distribution by region")
plt.pie(region_sales["total_sales"],labels=["East","North","South","West"],autopct="%.2f%%")

plt.show()

```

```python
HD=sale[sale["product"] == "Headphones"]["total_sales"].values
MT=sale[sale["product"] == "Monitor"]["total_sales"].values
LT=sale[sale["product"] == "Laptop"]["total_sales"].values
SW =sale[sale["product"] == "Smartwatch"]["total_sales"].values
TB=sale[sale["product"] == "Tablet"]["total_sales"].values

PD=[HD,MT,LT,SW,TB]

plt.title(" Sales distribution across products")

plt.boxplot(PD,tick_labels=label)

plt.yticks([x for x in range(1000,9000,1000)])
plt.ylabel("total product sales")
plt.xlabel("product sold")

plt.show()
```

```python
Gcustomer=sale.groupby("customer_id")["total_sales"].sum().reset_index()
```

```python
Bspend=Gcustomer[Gcustomer['total_sales'] >= 25000][["customer_id"]]
date_buyer=sale.groupby("customer_id")["monthly_sales"].count().reset_index()
```

```python
Fbuyer=date_buyer[date_buyer["monthly_sales"] >= 7 ][["customer_id"]]
Ncustomer=date_buyer[date_buyer["monthly_sales"] <= 4 ][["customer_id"]]

```

```python
Bspend['customer_type']="Big Spender"
Fbuyer["customer_type"] = "Frequent Buyer"
Ncustomer["customer_type"] = "New Customer"


CLV=pd.concat([Bspend,Fbuyer,Ncustomer],ignore_index=True)
CLV
```
