## CUSTOMIXE COLUMNS USING DEF

```python
def weight_category(x):
    if x["height_cm"] < 175 and x["weight_kg"] < 70:
        return "ligthweight"
    elif x["height_cm"] < 185 and x["weight_kg"] <= 80:
        return "middleweight"
    else:
        return "heavyweight"
bios_new["weigth_category"]=bios_new.apply(weight_category,axis=1)
bios_new[["name","weigth_category"]].head(25).query("weigth_category == 'heavyweight' ")
bios_new.head(6)
```

## HANDILING NULL VALUE
turning thing to non,
how to count value,
how to fill using fillna

```python
# turn into NAn
coffe.loc[[0,3],"Units Sold"]=np.nan
# how to know how many nan value you have
coffe.isna().sum()
```

```python
# how to fill any value 
coffe.loc[3:6,"Units Sold"]=15
coffe
```

```python
# fill nan value by average
coffe.fillna(coffe["Units Sold"].mean())
```

```python
# fill nan value by interpolate which fill nan base on surrounding data
coffe["Units Sold"]=coffe["Units Sold"].interpolate()
coffe
```

```python
coffe.loc[3:6,"Units Sold"]=np.nan
coffe
```

```python
# dropna drops nan value but using this in dangeruos because it delete the how index
#so use sub set to be specific
coffe.dropna(subset=["Units Sold"],inplace=False)
coffe
```

```python
# find row that is only nana and not NA
coffe[coffe["Units Sold"].isna()] # can use thos so that easiear to use fill 
coffe[coffe["Units Sold"].notna()]
```

## ADVANCE FUCNTIONALITY
.shift() .rank()  .cumsum() .rolling()

```python
coffe["yesterday_revenue"]=coffe["revenue"].shift(2) # shift 2 because there is 2 sama day
coffe["percent increase"]=coffe["revenue"] / coffe["yesterday_revenue"] * 100
coffe     
```

```python
bios_new["weight_rank"]=bios_new["weight_kg"].rank()
bios_new.sort_values(["weight_rank"],ascending=False)
```

```python
bios_new["height_rank"]=bios_new["height_cm" ].rank()
bios_new.sort_values(["height_rank"],ascending=False)
```

```python
latte=coffe[coffe["Coffee Type"] == "Latte"].copy()

latte["3day latte"]=latte["Units Sold"].rolling(3).sum()
latte.head(6)
```

```python
coffe["cumulative revenue"]=coffe["revenue"].cumsum()
coffe
```
